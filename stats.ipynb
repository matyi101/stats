```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import seaborn as sns
import numpy as np
import scipy as sp
import matplotlib.colors as mcolors
import scipy.stats as stats
sns.set()

plt.style.use('fivethirtyeight')
```

## 1. Exploratory Data Analysis



### 1.1 Islamic Banks - Loan Application by Purpose

```python
url = 'https://raw.githubusercontent.com/matyi101/stats/main/islamic_loan.csv'
df = pd.read_csv(url)
df.drop('consumer loan', inplace=True, axis=1)
df.set_index('date', inplace=True )
df.describe() 
```

```python
df.skew()
```

```python
df.plot(figsize=(10, 7), title ="Islamic Banks - Loan Application By Purpose")
```

```python
# Histogram

plt.figure(figsize=(15, 3))

plt.subplot(1, 4, 1)
df['vehicle'].hist(bins = 30, color="r")
plt.title('Vehicle')

plt.subplot(1, 4, 2)
df['property'].hist(bins = 30, color='b')
plt.title('Property')

plt.subplot(1, 4, 3)
df['personal'].hist(bins = 30, color='g')
plt.title('Personal Loan')

plt.subplot(1, 4, 4)
df['credit card'].hist(bins = 30, color='teal')
plt.title('Credit Cards')


# Density Distribution Plot

plt.figure(figsize=(15, 3))

plt.subplot(1, 4, 1)
sns.kdeplot(df['vehicle'], shade=True, color="r")
plt.title('Vehicle')

plt.subplot(1, 4, 2)
sns.kdeplot(df['property'], shade=True, color="b")
plt.title('Property')

plt.subplot(1, 4, 3)
sns.kdeplot(df['personal'], shade=True, color="g")
plt.title('Personal')

plt.subplot(1, 4, 4)
sns.kdeplot(df['credit card'], shade=True, color="teal")
plt.title('Credit Card')


plt.show()

# Q-Q Plot

plt.figure(figsize=(15, 3))

plt.subplot(1, 4, 1)

stats.probplot(df['vehicle'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")

plt.subplot(1, 4, 2)

stats.probplot(df['property'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")

plt.subplot(1, 4, 3)
stats.probplot(df['personal'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")


plt.subplot(1, 4, 4)
stats.probplot(df['credit card'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
```

## 1.2 Commercial Banks - Loan Applications by Purpose

```python
url = 'https://raw.githubusercontent.com/matyi101/stats/main/commercial_loan.csv'
df1 = pd.read_csv(url)
df1.drop('consumer loan', inplace=True, axis=1)
df1.set_index('date', inplace=True )
df1.describe()  
```

```python
# Histogram

plt.figure(figsize=(15, 3))

plt.subplot(1, 4, 1)
df1['vehicle'].hist(bins = 30, color="r")
plt.title('Vehicle')

plt.subplot(1, 4, 2)
df1['property'].hist(bins = 30, color='b')
plt.title('Property')

plt.subplot(1, 4, 3)
df1['personal'].hist(bins = 30, color='g')
plt.title('Personal Loan')

plt.subplot(1, 4, 4)
df1['credit card'].hist(bins = 30, color='teal')
plt.title('Credit Cards')


# Density Distribution Plot

plt.figure(figsize=(15, 3))

plt.subplot(1, 4, 1)
sns.kdeplot(df1['vehicle'], shade=True, color="r")
plt.title('Vehicle')

plt.subplot(1, 4, 2)
sns.kdeplot(df1['property'], shade=True, color="b")
plt.title('Property')

plt.subplot(1, 4, 3)
sns.kdeplot(df1['personal'], shade=True, color="g")
plt.title('Personal')

plt.subplot(1, 4, 4)
sns.kdeplot(df1['credit card'], shade=True, color="teal")
plt.title('Credit Card')


plt.show()

# Q-Q Plot

plt.figure(figsize=(15, 3))

plt.subplot(1, 4, 1)

stats.probplot(df1['vehicle'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")

plt.subplot(1, 4, 2)

stats.probplot(df1['property'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")

plt.subplot(1, 4, 3)
stats.probplot(df1['personal'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")


plt.subplot(1, 4, 4)
stats.probplot(df1['credit card'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
```
